```python
#Sort a list of strings by the length of each string from shortest to longest.
d=['kjkjkkj','aa','dbdbbdbd','lalalalal','oo']
new_d=sorted(d,key=lambda x:len(x))
print(new_d)
```

```python
# Sort a list of strings by the length of each string from longest to shortest
d=['kjkjkkj','aa','dbdbbdbd','lalalalal','oo']
new_d=sorted(d,key=lambda x:-len(x))
print(new_d)
```

```python
#sort the list by First element in the list
e=[[2,1],[1,2],[3,5],[4,3]]
new_e=sorted(e,key=lambda x:x[0])
print(new_e)
```
